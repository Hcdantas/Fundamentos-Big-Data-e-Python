# Acessando os elementos atraves do indice

Se nada mais for especificado, os valores serão rotulados com seu número de índice. O primeiro valor tem índice 0, o segundo valor tem índice 1 etc. Este rotulo pode ser usado para acessar um valor especifico.

```python
# Acessando o primeiro elemento
idades[0]

# Acessando o 4° elemento da Serie 
idades[3]
```

# Criando seus próprios índices

com o index argumento, você pode nomear seus próprios rótulos.

```python
# Criando uma nova lista

a = [1,7,2]

# Criar uma Series em função da Lista com índices personalizados

mySeries = pd.Series(a, index=['A','B','C'])

print(mySeries)
```

```python
# Criando uma lista para ser utilizada no índice

indice = ["Ana", "João" , "Maria"]

#Criar uma lista para ser utilizada nos valores da Series

notas = [10, 9, 8]

#Criar uma Series com a lista notas e com 
#índice de acordo cim a lista index

notasFinais = pd.Series(notas, index= indice)

print(notasFinais)
```

```python
# Criando um dicionário
calories = {
    'day1': 420,
    'day2': 380,
    'day3': 390
}
# Criar uma Série em função de um dicionálrio
myList = pd.Series(calories)

print(myList)
```
