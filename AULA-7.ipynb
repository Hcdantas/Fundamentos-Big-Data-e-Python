# Gere um arquivo em Excel para representar as vendas efetuadas pela filial Sudeste em 2010. Sendo que no relatório deve conter apenas vendas de casas sem garagem.

```python
df_final = tabela[
                    (tabela['filial']=='Sudeste') & 
                    (tabela['data']>='2010-01-01') & 
                    (tabela['data']<='2010-12-31') &
                    (tabela['tipo'] == 'Casa') &
                    (tabela['garage'] == 0)
                    ]

# Salvar em um diretório específico

caminho_destino = 'C:\\Users\\FIC\\Desktop\\24.10\\filiais\\'
arquivo_final = 'relatório-vendas-sudeste-2010-sem-garag.xlsx'

df_final.to_excel(caminho_destino + arquivo_final, sheet_name='EXEMPLO', index= False)            

```
